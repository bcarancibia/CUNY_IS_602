# Honduras Foreign Aid Project Locations

This is an exploratory data analysis of two datasets from the Government of Honduras (GoH) related to the amount of foreign financial aid flowing into the country. The two datasets are 1) the total amount of commitments for foreign financial aid in Honduras and 2) the locations of the financial aid in Honduras. Commitments are defined as the amount of money that Cooperation Donors i.e. World Bank, USAID, or the United Nations Development Programme (UNDP) and provide to countries for development. Commitments are a monetary number in USD that is typically used for analysis in the International Development Sector.  

The Honduras Foreign Aid Commitments was gathered and entered into the AMP by the Government of Honduras, the AMP data is stored in a PostgreSQL Database. The locations data for each project was obtained from the Donors that fund projects in Honduras. This location data, until now, is not shared with the GoH and is one of the many pitfalls of the international development sector.

```python
#import necessary libraries
import csv
import re
values = []
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
import sys
```

The data was obtained from two different sources. The Honduras Foreign Aid Projects are located in a PostgreSQL Database. The data was exported into a CSV from a Postgres view to ensure it was the most up to date data available. The other dataset (Honduras Locations) was downloaded in CSV format. The locations data was downloaded from the GoH AMP Website (www.pgc.sre.gob.hn). 

The two datasets were then imported into memory.

```python
#Read in the honduras commitments
df_commitments = pd.read_excel("/users/bcarancibia/CUNY_IS_602/Project/Commitments_Honduras.xlsx")
df_commitments.head()
```

```python
#read in the Honduras Locations data
df_locations = pd.read_excel("/users/bcarancibia/CUNY_IS_602/Project/Locations_Honduras.xlsx")
df_locations.head()
```

The projects and locations need to be joined by the ID known as Codigo PGC, which is the project id. After joining these two files together a new file will be created with just the columns needed to do the analysis.

```python
#merge the data
df_commitments_locations = df_locations.merge(df_commitments, on="Codigo PGC", how="outer")
df_commitments_locations.head()
```
