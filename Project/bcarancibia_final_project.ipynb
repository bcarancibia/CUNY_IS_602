# Honduras Foreign Aid Project Locations

This is an exploratory data analysis of two datasets from the Government of Honduras (GoH) related to the amount of foreign financial aid flowing into the country. The two datasets are 1) the total amount of commitments for foreign financial aid in Honduras and 2) the locations of the financial aid in Honduras. Commitments are defined as the amount of money that Cooperation Donors i.e. World Bank, USAID, or the United Nations Development Programme (UNDP) and provide to countries for development. Commitments are a monetary number in USD that is typically used for analysis in the International Development Sector.  

The Honduras Foreign Aid Commitments was gathered and entered into the AMP by the Government of Honduras, the AMP data is stored in a PostgreSQL Database. The locations data for each project was obtained from the Donors that fund projects in Honduras. This location data, until now, has never been shared with the GoH eliminating one of the many pitfalls of the international development sector.

```python
#import necessary libraries
import csv
import re
values = []
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
import sys
```

The data was obtained from two different sources. The Honduras Foreign Aid Projects are located in a PostgreSQL Database. The data was exported into a CSV from a Postgres view to ensure it was the most up to date data available. The other dataset (Honduras Locations) was downloaded in CSV format. The locations data was downloaded from the GoH AMP Website (www.pgc.sre.gob.hn). 

The two datasets were then imported into memory.

```python
#Read in the honduras commitments
df_commitments = pd.read_excel("/users/bcarancibia/CUNY_IS_602/Project/Commitments_Honduras.xlsx")
df_commitments.head()
```

```python
#read in the Honduras Locations data
df_locations = pd.read_excel("/users/bcarancibia/CUNY_IS_602/Project/Locations_Honduras.xlsx")
df_locations.head()
```

The projects and locations need to be joined by the ID known as Codigo PGC, which is the project id. After joining these two files together a new file will be created with just the columns needed to do the analysis.

```python
#merge the data
df_commitments_locations = df_locations.merge(df_commitments, on="Codigo PGC", how="outer")
df_commitments_locations.head()
```

The average project funded in Honduras is  17,065,300 USD with the standard deviation being 20,776,332 USD. Statistically, a standard deviation that is larger than a mean indicates that there will be little to no statistically significant differences between mean and zero. This means that the data is non-normal. This is acceptable for the dataset because foreign aid projects are not tested if they are significant based t-tests. What determine a foreign aid projects significance is based on outcome of the projects i.e. municipality's poverty rate went down. 

```python
df_commitments_locations.Total.plot()
```

As seen in ordered by aggregateed sector (below) there are 19 different sectors with the largest amount of foreign aid going to Multisector and the smallest amount of aid going to Programas y Politicas de Poblacion y Salud (Programs and Politics of Health and Population). The most specific sectors and receive the most amount of aid are Transporte y Almacenaje and Agriculture. This makes sense because Hurricane Mitch in 1998 wiped out about 80% of Honduras' infrasture and the country is still trying to recover from that hurricane. Also an interesting thing to look at in the data is that there is a sector without a named defined. There still obvious data clean up and better data management that needs to be done in-country.
