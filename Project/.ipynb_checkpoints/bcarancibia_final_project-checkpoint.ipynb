# Honduras Foreign Aid Project Locations

This is an analysis and data clean up of two datasets from the Government of Honduras related to the country's foreign financial aid. The two datasets are 1) the total amount of commitments for foreign financial aid in Honduras and 2) the locations of the financial aid in Honduras. Commitments are defined as the amount of money that Cooperation Donors i.e. World Bank, USAID, or the United Nations Development Programme (UNDP) and provide to countries for development. Commitments are a monetary number in USD that is typically used for analysis in the International Development Sector.  

```python
import csv
import re
values = []
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
df_commitments = pd.read_excel("/users/bcarancibia/CUNY_IS_602/Project/Commitments_Honduras.xlsx")
df_commitments.head()
```

```python
df_locations = pd.read_excel("/users/bcarancibia/CUNY_IS_602/Project/Locations_Honduras.xlsx")
df_locations.head()
```

```python
df_commitments_locations = df_locations.merge(df_commitments, on="Codigo PGC", how="outer")
df_commitments_locations.head()
```
